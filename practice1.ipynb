# Histograms

```python
sns.distplot(iris_df['sepal_length'], label='sepal length', color='b')
sns.distplot(iris_df['sepal_width'], label='sepal length', color='r')
plt.legend()
plt.show();
```

```python
sns.distplot(iris_df['sepal_length'], bins=20, label='sepal length', color='b')
sns.distplot(iris_df['sepal_width'], bins=20, label='sepal length', color='r')
plt.legend()
plt.show();
```

```python
# only distribution
sns.kdeplot(iris_df['sepal_length'], shade=True,label='sepal length', color='r')
sns.kdeplot(iris_df['sepal_width'], shade=True,label='sepal width', color='b')
plt.legend()
```

```python
# subplots
# share x- xticks will be shared.
f, axes= plt.subplots(2,2, figsize=(7,7), sharex=True)
sns.distplot(iris_df['sepal_length'], bins=20, label='sepal length', color='b', ax= axes[0,0])
sns.distplot(iris_df['sepal_width'], bins=20, label='sepal length', color='r', ax= axes[0,1])
sns.distplot(iris_df['petal_length'], bins=20, label='petal_length', color='b', ax= axes[1,0])
sns.distplot(iris_df['petal_width'], bins=20, label='petal_width', color='r', ax= axes[1,1])
plt.show();
```

# Distribution- joint distributions
Can be used to show correlation between two variables

```python
#  There are different types of graph you can choose:
# o scatter, kde, hex, resid, reg, etc
sns.jointplot(x=iris_df['sepal_length'], y= iris_df['sepal_width'], kind= 'scatter' ,color= 'r')
plt.show();
```

```python
sns.jointplot(x=iris_df['sepal_length'], y= iris_df['sepal_width'], kind= 'kde' ,color= 'r')
plt.show();
```

#### Scatter Plot- RegPlot

```python
sns.regplot(x=iris_df['petal_length'], y=iris_df['petal_width'], color='r')
plt.show();
```

#### ScatterPlot- lmplot
Advanced version of the regplot. gives hue and markers and other useful options


```python
sns.lmplot(x='sepal_length', y= 'sepal_width', data= iris_df, fit_reg=False, hue= 'species', legend= True, markers=['o','x', '1'])
plt.plot();
```

##### Bar Chart

```python
# Histograms are used for continous non categorical data for example, length of a petal. 
# Bar chart are used for categorical data.This can be countable and can be conpared to next bar.
#For example: tips for everyday in a week. here aswell there is a grouping being performed.

# petal length cannot be compared exactly because there will be lot of petal.
# If we need to compare, we need to perform some kind of grouping based on the species.
# tips and total on the other hand should be grouped based on the day of the week which can be compared to other days.
tips= sns.load_dataset('tips')
print(tips.head())
ax= sns.barplot(x='day', y='total_bill', data= tips)
plt.show();

# The plot shows the average bill. The line at the top shows the confidence interval(standard deviation)
```

```python
# addng a thid category for comparision using hue
ax= sns.barplot(x='day', y='total_bill',hue='sex', data= tips)
plt.show();
```
